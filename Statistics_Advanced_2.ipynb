Q1. What is hypothesis testing in statistics

 Hypothesis testing is a statistical method used to evaluate assumptions about a population using sample data. It determines if an observed result is a true pattern or merely a random occurrence, allowing researchers to make data-driven decisions.The process follows a structured four-step framework:    
**1. State the Hypotheses**
 Every test requires two competing assumptions:
 *  Null Hypothesis (H₀): The default assumption that there is no effect, no difference, or no relationship. It represents the status quo (e.g., "The new medicine is no different than the old one").
*   Alternative Hypothesis (\(H_{A}\) or H₁): The claim you are testing for, suggesting an effect, difference, or relationship exists.

 **2. Choose a Significance Level (α)**
 The significance level is the probability threshold set before analyzing the data, typically set at 0.05 (5%). It determines how much evidence is needed to reject the null hypothesis.

 **3. Calculate the Test Statistic and p-Value** Using the sample data, you calculate a test statistic (such as a t-score or z-score) to determine how far your data deviates from what the null hypothesis expects. This generates a p-value, which measures the probability of getting your results assuming the null hypothesis is true.
  
 **4. Make a Decision** You compare the p-value to your significance level (α):
 * If the p-value ≤ α: You reject the null hypothesis. The results are considered statistically significant, meaning the evidence supports the alternative hypothesis.
 * If the p-value > α: You fail to reject the null hypothesis. There is not enough evidence to prove the alternative claim

What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis is the default assumption that there is no effect, no difference, or no relationship between variables in a population. It essentially states that any observed results are due to random chance. The alternative hypothesis is the researcher's claim, stating that a true effect, difference, or relationship does exist.
#Key Differences at a Glance
In hypothesis testing, the null and alternative hypotheses act as competing statements:
* Null Hypothesis ((H_{0})): Assumes "no change" or "no effect" (status quo). It always uses equality symbols . Researchers assume this is true until the data proves otherwise.
* Alternative Hypothesis ((H_{a}) or (H_{1})): Represents the researcher's specific claim or expectation. It uses inequality symbols ((=), (<), or (>)). Researchers aim to gather enough evidence to reject the null hypothesis in favor of this one.
#Example Scenario
Imagine you are testing a new medication for high blood pressure:
* Null Hypothesis ((H_{0})): The new medication has no effect on blood pressure; any changes are just random chance ((H_0:{Effect} = 0)).
* Alternative Hypothesis ((H_{a})): The new medication effectively lowers blood pressure ((H_a:{Effect} < 0))

Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

 The significance level (denoted as a) is the predetermined threshold for deciding whether to reject the null hypothesis. It represents the maximum probability you are willing to accept of incorrectly rejecting a true null hypothesis (a Type I error). Common values are0.05(5%) or 0.01(1%).  
 # Role in Deciding the Outcome

 In hypothesis testing, the significance level acts as a strict boundary line when analyzing your sample data. It determines the outcome of the test in the following way:

**1.Calculate the P-value:** After running a statistical test, you calculate a (p)-value, which measures the probability of obtaining your observed results (or more extreme results) assuming the null hypothesis is true.   
 **2.Compare P-value to (a):** You compare this (p)-value directly to your predetermined significance level (a).
 * If (p < a): The results are statistically significant. The probability of obtaining these results by pure chance is lower than your risk threshold. You reject the null hypothesis in favor of the alternative hypothesis.
 * If (p > a): The results are not statistically significant. There is insufficient evidence to conclude that a true effect or difference exists. You fail to reject the null hypothesis.Choosing the Significance LevelSetting the significance level requires balancing risk. A stricter significance level like (0.01) makes it harder to reject the null hypothesis, reducing false positives. Conversely, a higher level like (0.10) makes it easier to find an effect but increases the risk of a false positive. The choice depends on the field of study and the real-world consequences of making an incorrect decision
 For example, in medical research, a lower significance level (e.g(0.01)) is often preferred because an incorrect conclusion could lead to dangerous treatments.

Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

=The primary difference between a Z-test and a T-test is whether the population standard deviation is known. A Z-test uses the standard normal distribution, whereas a T-test relies on the Student's T-distribution, which features wider tails to account for the extra uncertainty of estimating parameters.
#When to Use a Z-Test
Use a Z-test when you want to compare a sample mean to a population mean (or compare two sample means) and the following conditions are met:
* Known Population Standard Deviation: The population standard deviation (\(\sigma \)) is already known.
* Large Sample Size: The sample size (\(n\)) is large, typically 30 or more, which invokes the Central Limit Theorem.
* Test for Proportions: Z-tests are also ideal for testing differences between proportions (e.g., A/B testing conversion rates) when expected successes and failures are adequately large.
#When to Use a T-Test
Use a T-test in similar scenarios, but when the following conditions differ:
* Unknown Population Standard Deviation: The population standard deviation is unknown, forcing you to estimate it using the sample standard deviation (\(s\)).
* Small Sample Size: The sample size is small (typically less than 30), meaning the T-distribution accurately compensates for the added uncertainty of smaller datasets.

Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Define binomial distribution parameters
n = 10          # Number of trials
p = 0.5         # Probability of success in each trial
size = 10000    # Number of data points to generate (sample size)

# Generate binomial distribution data
data = np.random.binomial(n=n, p=p, size=size)

# Set up the histogram bins aligned perfectly with discrete integer values
bins = np.arange(0, n + 2) - 0.5

# Plot the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=bins, rwidth=0.8, color='skyblue', edgecolor='black', density=True)

# Add titles and axis labels
plt.title(f'Binomial Distribution Histogram (n={n}, p={p})', fontsize=14)
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Density / Probability', fontsize=12)

# Ensure all discrete outcomes (0 to 10) are shown on the x-axis
plt.xticks(range(n + 1))

# Display the grid lines for readability
plt.grid(axis='y', alpha=0.75, linestyle='--')

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest

# Provided dataset (Sample size n = 36)
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# 1. Define known parameters
mu_0 = 50.0          # Hypothesized population mean
sigma = 0.5          # Known population standard deviation
alpha = 0.05         # Significance level

# 2. Manual Z-test derivation
n = len(sample_data)
sample_mean = np.mean(sample_data)
standard_error = sigma / np.sqrt(n)

z_stat = (sample_mean - mu_0) / standard_error
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("--- Manual Method (Known Sigma) ---")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value:     {p_value:.4f}")

# 3. Quick confirmation using statsmodels
# Note: statsmodels uses sample standard deviation (s) internally
z_stat_sm, p_val_sm = ztest(sample_data, value=mu_0)
print("\n--- Statsmodels Method (Sample Std Dev Estimator) ---")
print(f"Z-statistic: {z_stat_sm:.4f}")
print(f"P-value:     {p_val_sm:.4f}")
```

 Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

#Visual and Functional Breakdown
**Data Generation:** The np.random.normal function draws random parameters matching your target population bounds.    
**Confidence Interval Calculation:** We utilize stats.t.interval from SciPy Stats Suite. The Student's t-distribution is selected instead of a Z-score distribution since the underlying population parameters are calculated dynamically from the sample.      
**Standard Error:** The bounds scale inversely with the square root of the sample size (np.sqrt(sample_size_n)).     
**Visual Highlights:** plt.axvline shows the sample mean center point while plt.axvspan creates a highlighted shaded range explicitly framing the calculated lower and upper bounds of your confidence interval

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Simulate data from a normal distribution
np.random.seed(42)  # Set seed for reproducible results
true_mean = 50
true_std = 10
sample_size = 250

# Generate the random samples
data = np.random.normal(loc=true_mean, scale=true_std, size=sample_size)

# 2. Calculate sample statistics
sample_mean = np.mean(data)
sample_size_n = len(data)
# Use ddof=1 for the sample standard deviation (unbiased estimator)
sample_std = np.standard_deviation = np.std(data, ddof=1)
# Calculate standard error of the mean (SEM)
standard_error = sample_std / np.sqrt(sample_size_n)

# 3. Calculate 95% confidence interval using Student's t-distribution
# (Best practice when population standard deviation is unknown)
confidence_level = 0.95
degrees_of_freedom = sample_size_n - 1

ci_lower, ci_upper = stats.t.interval(
    confidence_level,
    df=degrees_of_freedom,
    loc=sample_mean,
    scale=standard_error
)

# Print out the computed values
print(f"True Mean: {true_mean}")
print(f"Sample Mean: {sample_mean:.3f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.3f}, {ci_upper:.3f})")

# 4. Plot the data using Matplotlib
plt.figure(figsize=(10, 6))

# Plot the histogram of simulated data
count, bins, ignored = plt.hist(
    data,
    bins=20,
    density=True,
    alpha=0.6,
    color='skyblue',
    edgecolor='black',
    label='Simulated Data Histogram'
)

# Plot the theoretical probability density function (PDF) curve
xmin, xmax = plt.xlim()
x_axis = np.linspace(xmin, xmax, 100)
plt.plot(
    x_axis,
    stats.norm.pdf(x_axis, sample_mean, sample_std),
    color='navy',
    linestyle='--',
    linewidth=2,
    label='Fitted Normal Curve'
)

# Visual anchors for the sample mean and its 95% Confidence Interval
plt.axvline(
    sample_mean,
    color='crimson',
    linestyle='-',
    linewidth=2.5,
    label=f'Sample Mean ({sample_mean:.2f})'
)
plt.axvspan(
    ci_lower,
    ci_upper,
    color='yellow',
    alpha=0.3,
    label=f'95% Mean CI ({ci_lower:.2f} to {ci_upper:.2f})'
)

# Final plot formatting
plt.title('Simulated Normal Distribution & 95% Confidence Interval of the Mean', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.3)

# Display the plot
plt.tight_layout()
plt.show()
```

Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
  
To calculate Z-scores and visualize the standardized distribution, you can use the scipy.stats.zscore function to compute the values and matplotlib.pyplot.hist to plot the resulting histogram.

A Z-score represents how many standard deviations a specific data point lies away from the mean of its dataset

#The Empirical Rule (68-95-99.7 Rule)
If your original data follows a normal (bell-shaped) curve, its Z-scores will adhere to these statistical benchmarks:   
~ 68% of all data points will have a Z-score between -1 and +1.   
~ 95% of all data points will have a Z-score between -2 and +2.    
~ 99.7% of all data points will have a Z-score between -3 and +3.   
 Any value falling outside this range (|Z|>3) is generally flagged as a statistical

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_visualize_zscores(data=None):
    """
    Calculates Z-scores for a dataset and plots a histogram of the standardized data.
    """
    # 1. Generate sample data if none is provided
    if data is None:
        # Generating 1000 points from a distribution with mean=50 and std=10
        np.random.seed(42)
        data = np.random.normal(loc=50, scale=10, size=1000)
    else:
        data = np.array(data)

    # 2. Calculate Z-scores
    # Formula applied under the hood: Z = (X - mean) / std
    z_scores = stats.zscore(data)

    # 3. Create the histogram visualization
    plt.figure(figsize=(8, 5))

    # Plotting the standardized data
    plt.hist(z_scores, bins=30, edgecolor='black', color='#3498db', alpha=0.7)

    # Add visual indicators for the mean and standard deviations
    plt.axvline(x=0, color='red', linestyle='--', linewidth=2, label='Mean (Z = 0)')
    plt.axvline(x=1, color='orange', linestyle=':', linewidth=1.5, label='±1 Std Dev')
    plt.axvline(x=-1, color='orange', linestyle=':')
    plt.axvline(x=2, color='green', linestyle=':', linewidth=1.5, label='±2 Std Dev')
    plt.axvline(x=-2, color='green', linestyle=':')

    # Styling and labeling the plot
    plt.title('Histogram of Standardized Data (Z-Scores)', fontsize=14)
    plt.xlabel('Z-Score (Standard Deviations from Mean)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)

    # Display the plot
    plt.show()

    return z_scores

# Execute the function
z_results = calculate_and_visualize_zscores()
```
